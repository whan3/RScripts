```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from scipy.stats import ttest_rel
from statsmodels.stats.multitest import multipletests
import os

# Directory where datasets are stored
directory = "E:/R(5)/Datasets/"

# Classifiers
classifiers = {
    "RF": RandomForestClassifier(),
    "GB": GradientBoostingClassifier(),
    "AdaB": AdaBoostClassifier(),
    "XgB": XGBClassifier(),
    "DT": DecisionTreeClassifier(),
    "SVM": SVC(),
    "NB": GaussianNB()
}

# Function to evaluate a classifier
def evaluate_classifier(clf, X_train, X_test, y_train, y_test):
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    return accuracy_score(y_test, predictions)

# Initialize a DataFrame to store results
results_df = pd.DataFrame(columns=['Dataset', 'Classifier 1', 'Classifier 2', 'P-Value', 'Adjusted P-Value'])

# Process each dataset
for i in range(1, 8):
    file_path = os.path.join(directory, f"Dataset_{i}.csv")
    df = pd.read_csv(file_path)
    
    X = df.iloc[:, :-1]  # Features
    y = df.iloc[:, -1]   # Target

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Evaluate each classifier
    accuracies = {}
    for clf_name, clf in classifiers.items():
        accuracies[clf_name] = evaluate_classifier(clf, X_train, X_test, y_train, y_test)

    # Perform paired t-tests
    for clf1 in classifiers:
        for clf2 in classifiers:
            if clf1 != clf2:
                stat, p = ttest_rel([accuracies[clf1]], [accuracies[clf2]])
                results_df = results_df.append({
                    'Dataset': f'Dataset{i}',
                    'Classifier 1': clf1,
                    'Classifier 2': clf2,
                    'P-Value': p
                }, ignore_index=True)

# Adjust for multiple comparisons
p_values = results_df['P-Value'].values
adjusted_p_values = multipletests(p_values, method='bonferroni')[1]
results_df['Adjusted P-Value'] = adjusted_p_values
```
