```python
import pandas as pd
df  = pd.read_csv('E:/R(5)/kidney_disease.csv')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = df.applymap(lambda x: x.replace(" ", "").replace("\t", "") if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace(" ", "").replace("?", "") if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace(" ", "").replace("\t", "") if isinstance(x, str) else x)
df.replace("", pd.NA, inplace=True)
df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

columns_to_encode = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']

for column in columns_to_encode:
    mask_not_null = ~df[column].isnull()
    df.loc[mask_not_null, f'{column}_le'] = label_encoder.fit_transform(df.loc[mask_not_null, column])
df.head()
```

```python
df['classification'].replace({"ckd":"1", "notckd":"0" }, inplace= True)
df
```

```python
df1 = df.drop(['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane'], axis=1)
df1
```

```python
import pandas as pd

columns_with_nulls1 = ['sg', 'al', 'su', 'rbc_le', 'pc_le', 'pcc_le', 'ba_le', 'htn_le', 'dm_le', 'cad_le', 'appet_le', 'pe_le', 'ane_le']

columns_with_nulls2 = ['age', 'bp',  'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo',  'pcv', 'wc', 'rc']



for column1 in columns_with_nulls1:
    mode_val = df1[column1].mode()[0]  # Getting the mode value of the column
    df1[column1].fillna(mode_val, inplace=True)
    
for column2 in columns_with_nulls2:
    mean_val = df1[column2].mean()  # Getting the mode value of the column
    df1[column2].fillna(mean_val, inplace=True)
    
df1
```
